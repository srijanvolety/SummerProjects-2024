```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import eig

N = 100
d = 2 * np.pi / 99

k_ = np.arange(-np.pi, np.pi, 0.01)



x_values = np.linspace(-np.pi,np.pi,num=N) #100x100 matrix
V_values = np.sin(x_values)
V = np.diag(V_values)

#  kinetic energy
T1 = np.zeros((N, N), dtype=complex)
for l in range(N):
  for m in range(N):
    if l == m:
      T1[l, m] = 2
    elif m == l - 1:
      T1[l, m] = -1
    elif m == l + 1:
      T1[l, m] = -1

# Make the Hamiltonian matrix periodic
T1[0,99]=-1
T1[-1,0] =-1

# Compute the kinetic term
T1=(1/(d**2))*T1

#2pq matrix

T2 = np.zeros((N, N), dtype=complex)
for l in range(N):
  for m in range(N):
    if l == m:
     T2[l, m] =0
    elif m == l - 1:
     T2[l, m] = 1j
    elif m == l + 1:
     T2[l, m] = -1j


T2[0,99]=1j
T2[-1,0] =-1j
T2=  (1/d)*T2

#q^2matrix

T3 = np.identity(100)



'''#Hamiltonian
H=(T1+T2+T3)+V'''

band_1 = []
band_2 = []
band_3 = []
band_4 = []
band_5 = []
band_6 = []
band_7 = []
band_8 = []
band_9 = []
band_10 = []
for q in k_:
  H= T1+T2*q +(q**2)*T3 + V
  u, _ =eig(H)       #u=eigenvalue v=eigenvec
  u_sorted = sorted(u)




# Store eigenvalues in bands
  band_1.append(u_sorted[0])
  band_2.append(u_sorted[1])
  band_3.append(u_sorted[2])
  band_4.append(u_sorted[3])


plt.plot(k_, band_1)
plt.plot(k_, band_2)
plt.plot(k_, band_3)

plt.xlabel('k')
plt.ylabel('Energy')
plt.title('Energy Bands')
plt.xlim(-0.5,0.5)

plt.show()

for i in band_1:
      plt.axhline(y = i)
for i in band_2:
      plt.axhline(y = i)
for i in band_3:
      plt.axhline(y = i)
plt.show()


















```
