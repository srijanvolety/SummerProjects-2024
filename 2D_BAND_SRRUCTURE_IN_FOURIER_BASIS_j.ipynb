```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import eig
from numpy.linalg import eig
from mpl_toolkits.mplot3d import Axes3D



# Initialize lists to store q_x, q_y, and eigenvalues
q_xlist = []
q_ylist = []
eigenvalues_list = []

g=2*np.pi
# Loop over q_x and q_y values
for q_x in np.arange(-np.pi, np.pi, 0.01):
    for q_y in np.arange(-np.pi, np.pi, 0.01):
        # Define the matrix size
        N = 9

        # Initialize the Hamiltonian matrix with zeros
        H = np.zeros((N, N), dtype=complex)

        # Assign values to the Hamiltonian matrix
        for i in range(N):
            for j in range(N):
                if i == 0 and j == 0:
                    H[i, j] = (q_x+g)**2 + (q_y + g)**2
                elif i == 1 and j == 1:
                    H[i, j] = (q_x )**2 + (q_y +g)**2
                elif i == 2 and j == 2:
                    H[i, j] = (q_x - g)**2 + (q_y+g)**2
                elif i == 3 and j == 3:
                    H[i, j] = (q_x + g)**2 + (q_y)**2
                elif i == 4 and j == 4:
                    H[i, j] = (q_x**2) + (q_y )**2
                elif i == 5 and j == 5:
                    H[i, j] = (q_x - g)**2 + (q_y)**2
                elif i == 6 and j == 6:
                    H[i, j] = (q_x + g)**2 + (q_y - g)**2
                elif i == 7 and j == 7:
                    H[i, j] = (q_x )**2 + (q_y - g)**2
                elif i == 8 and j == 8:
                    H[i, j] = (q_x - g)**2 + (q_y - g)**2
                elif i==0 and j==4:
                    H[i, j] =0.5
                elif i==1 and j==3:
                    H[i, j] =0.5
                elif i==1 and j==5:
                    H[i, j] =0.5
                elif i==2 and j==4:
                    H[i, j] =0.5
                elif i==6 and j==4:
                    H[i, j] =0.5
                elif i==8 and j==4:
                    H[i, j] =0.5
                elif i==7 and j==3:
                    H[i, j] =0.5
                elif i==3 and j==1:
                    H[i, j] =0.5
                elif i==5 and j==1:
                    H[i, j] =0.5
                elif i==4 and j==0:
                    H[i, j] =0.5
                elif i==4 and j==2:
                    H[i, j] =0.5
                elif i==4 and j==6:
                    H[i, j] =0.5
                elif i==4 and j==8:
                    H[i, j] =0.5
                elif i==3 and j==7:
                    H[i, j] =0.5
                elif i==5 and j==7:
                    H[i, j] =0.5
                elif i==7 and j==5:
                    H[i, j] =0.5

        # Calculate eigenvalues and eigenvectors
        u, v = eig(H)
        u=np.real(u)

        # Sort eigenvalues
        u.sort()

        '''v = [u[i] for i in range(2)]'''
             # Store q_x, q_y, and eigenvalues for this iteration
        q_xlist.extend([q_x] * len(u))
        q_ylist.extend([q_y] * len(u))
        eigenvalues_list.extend(u)

# Convert lists to numpy arrays for plotting
qx_array = np.array(q_xlist)
qy_array = np.array(q_ylist)
eigenvalues_array = np.array((eigenvalues_list))

# Plotting the 3D plot

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

    # Scatter plot for each eigenvalue
sc = ax.scatter(qx_array, qy_array, eigenvalues_array, c=eigenvalues_array, cmap='viridis')

# Set labels
ax.set_xlabel('qx')
ax.set_ylabel('qy')
ax.set_zlabel('Eigenvalues')




# Add a color bar which maps values to colors
plt.colorbar(sc)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import eig

# Parameters
N = 100
q_x = np.linspace(-0.5, 0.5, N)
q_y = np.linspace(-0.5, 0.5, N)
# h = 0.00025

# Define the function H
def H(q_x, q_y):
    H = np.zeros((9, 9), dtype=complex)
    for i in range(9):
        for j in range(9):
            if i == 0 and j == 0:
                H[i, j] = (q_x + 1)**2 + (q_y + 1)**2
            elif i == 1 and j == 1:
                H[i, j] = (q_x)**2 + (q_y + 1)**2
            elif i == 2 and j == 2:
                H[i, j] = (q_x - 1)**2 + (q_y + 1)**2
            elif i == 3 and j == 3:
                H[i, j] = (q_x + 1)**2 + (q_y)**2
            elif i == 4 and j == 4:
                H[i, j] = (q_x**2) + (q_y)**2
            elif i == 5 and j == 5:
                H[i, j] = (q_x - 1)**2 + (q_y)**2
            elif i == 6 and j == 6:
                H[i, j] = (q_x + 1)**2 + (q_y - 1)**2
            elif i == 7 and j == 7:
                H[i, j] = (q_x)**2 + (q_y - 1)**2
            elif i == 8 and j == 8:
                H[i, j] = (q_x - 1)**2 + (q_y - 1)**2
            elif (i == 0 and j == 4) or (i == 1 and j == 3) or (i == 1 and j == 5) or (i == 2 and j == 4) or (i == 6 and j == 4) or (i == 8 and j == 4) or (i == 7 and j == 3) or (i == 3 and j == 1) or (i == 5 and j == 1) or (i == 4 and j == 0) or (i == 4 and j == 2) or (i == 4 and j == 6) or (i == 4 and j == 8) or (i == 3 and j == 7) or (i == 5 and j == 7) or (i == 7 and j == 5):
                H[i, j] = 1

    eigen_values, _ = eig(H)
    #eigen_values = np.real(np.abs(eigen_values))
    eigen_values.tolist()
    eigen_values.sort()

    return eigen_values

# Create the meshgrid
q_x_grid, q_y_grid = np.meshgrid(q_x, q_y)

# Compute eigenvalues for the meshgrid
mesh = np.zeros((len(q_x), len(q_y), 9))

for i in range(len(q_x)):
    for j in range(len(q_y)):
        mesh[i, j] = H(q_x[i], q_y[j])

# Plotting the surface plots for each eigenvalue
fig = plt.figure(figsize=(15, 15))

for i in range(9):
    ax = fig.add_subplot(3, 3, i + 1, projection='3d')
    ax.plot_surface(q_x_grid, q_y_grid, mesh[:, :, i], cmap='viridis')
    ax.set_title(f'E{i + 1}')
    ax.set_xlabel('$q_x$')
    ax.set_ylabel('$q_y$')
    ax.set_zlabel('$E$')

plt.tight_layout()
plt.show()
```
